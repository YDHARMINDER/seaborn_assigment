relplot:To make complex comparisons easier, Seaborn provides a function called relplot, which is short for relationship plot. relplot takes some of the same arguments as scatterplot, such as data, x, y, and hue, but adds other arguments as well.
jointplot:Joint plots allow you to create helpful visuals that plot both a bivariate distribution (such as a scatter plot), as well as the distribution of each of the individual variables.
pairplot:A pairplot plot a pairwise relationships in a dataset. The pairplot function creates a grid of Axes such that each variable in data will by shared in the y-axis across a single row and in the x-axis across a single column.
